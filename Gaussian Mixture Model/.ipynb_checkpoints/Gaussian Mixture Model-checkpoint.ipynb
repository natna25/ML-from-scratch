# Coding A Gaussian Mixture Model (GMM) from scratch


```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
```

```python
def gen_data(samples=300):
    # generate random sample, two components
    np.random.seed(0)
    # generate spherical data centered on (20, 20)


    shifted_gaussian = np.random.randn(samples, 2) + np.array([20, 20])
    # generate zero centered stretched Gaussian data
    C = np.array([[0., -0.7], [3.5, .7]])
    stretched_gaussian = np.dot(np.random.randn(samples, 2), C)
    # concatenate the two datasets into the final training set
    return np.vstack([shifted_gaussian, stretched_gaussian])

X_train = gen_data(samples=300)
```

## EM Algorithm: Expactation Maximization

In this problem type, it resembles somewhat the Gaussian Discriminant Analysis (GDA) that we did previously except for two differences:

1) It is clear that the two data distributions __do not have the same covariance__ 

2) The problem type is __unsupervised__



Like in all clustering problems, we have to choose the number of clusters that we wish to find. Usually you would experiment with this hyperparameter until you got the best results using grid search or any other hyperparameter searching technique. For us, we will simply choose a number of clusters as 2 from the begining.


### Initialize the parameters

```python
from sklearn.datasets import make_spd_matrix

def init_params(data,n_clusters=2):


    #mixture parameter
    mix = 0.5
    phi = np.array([mix,1-mix])
    #means
    mu = np.random.uniform(data.min() ,data.max() ,  size=(n_clusters,data.shape[1]) )
    #covariance matrices
    sigma_min=0.1
    sigma_max=1
    
    cov = []
    for k in range(n_clusters):
    
        cov.append(make_spd_matrix(data.shape[1]) )
    
    #convert to numpy array for simpler access
    cov  = np.array(cov)
    
    return phi,mu,cov
```

# E and M step

The EM algorithm consists in two distinct steps, the E step or expectation step where we update the value of our soft weight $w_j$ that serves as a proxy for the indicator function that we would normally use in the MLE optimization technique. Then comes the M step where for the currently given parameters, we try to maximize the lower bound likelyhood. This lets us proceed to another E step where we can further refine the values of $w_j$ and so on.

```python
from math import sqrt

def pdf(mu,sigma,x):
    
    #mu = mu.reshape((mu.shape[0],1))
    #x = x.reshape((x.shape[0],1))
    
    #A = 1/(np.power(2*np.pi ,sigma.shape[0]/2) * np.power(abs(np.linalg.det(sigma)),1/2))
    #B = np.exp((-1/2)* np.dot(np.dot(np.transpose(x - mu), np.linalg.inv(sigma) ), (x - mu) ) )
    
    frac = (2*np.pi)**(x.shape[0]/2)
    
    frac *=  np.power(abs(np.linalg.det(sigma)),1/2)
    
    frac = (1/frac)
    
    x_mu = x - mu
    inv_sigma = np.linalg.inv(sigma)
    
    expo = np.exp((-0.5)* np.dot(np.dot(x_mu.T, inv_sigma ),x_mu))
    
    
    """
    print(x - mu)
    print(np.linalg.inv(sigma))
    print(np.dot(np.transpose(x - mu), np.linalg.inv(sigma) ))
    print( (-1/2)*np.dot(np.dot(np.transpose(x - mu), np.linalg.inv(sigma) ), (x - mu) ))
    """
    return (frac*expo)
```

```python
def E_step(phi,mu,cov,clusters,data):
    epsilon = 1e-8
    
    w = np.zeros((data.shape[0],clusters))
    
    for i in range(data.shape[0]):
        
        for k in range(clusters):
            #find w
            w[i,k] = (pdf(mu[k],cov[k],data[i]) * phi[k] )/(sum([pdf(mu_l,cov_l,data[i])*phi_l for mu_l,cov_l,phi_l in zip(mu,cov,phi)])) 
            w[i,k] += epsilon
            
    return w
```

```python
def M_step(phi,mu,cov,clusters,data,w):
    
    phi = np.sum(w,axis=0)/w.shape[0]

    
    mu = np.zeros((clusters,data.shape[1])) 
    cov = np.zeros((clusters,data.shape[1],data.shape[1])) 
    
    for i,x in enumerate(data):
        #numerator sum
        mu[0] += w[i,0] * x
        
        mu[1] += w[i,1] * x
        
        
    
    #denominator sum
    mu[0] /= np.sum(w[:,0])
    mu[1] /= np.sum(w[:,1])
    
    
    for i,x in enumerate(data):
        x_mu_0 = (x-mu[0]).reshape((2,1))
        x_mu_1 = (x-mu[1]).reshape((2,1))
        
        
        cov[0] += w[i,0] * np.dot(x_mu_0, x_mu_0.T )
        cov[1] += w[i,1] * np.dot(x_mu_1, x_mu_1.T )
        
        
    
    cov[0] /= np.sum(w[:,0])
    cov[1] /= np.sum(w[:,1])
    
    
    print("phi",phi)
    print("mu",mu)
    print("cov",cov)
    
    
    return phi,mu,cov

```

```python
def plot_boundary(mu,cov,pos,X_train):

    colors = ['tab:blue', 'tab:orange', 'tab:green', 'magenta', 'yellow']

    plt.figure(figsize=(12,int(8)))

    likelyhood = []
    for k in range(n_clusters):

        likelyhood.append([pdf(mu[k],cov[k],np.array([p[0],p[1]])) for p in pos] )

    likelyhood = np.array(likelyhood)
    likelyhood.shape

    preds = np.argmax(likelyhood,axis=0)

    for c in range(n_clusters):

        pred_ids = np.where(preds == c)
        plt.scatter(pos[pred_ids[0],0], pos[pred_ids[0],1], color=colors[c], alpha=0.4, edgecolors='none', marker='s')


    plt.scatter(X_train[...,0], X_train[...,1], facecolors='none', edgecolors='grey')

    for c in range(n_clusters):
        plt.scatter(mu[c][0], mu[c][1], color=colors[c], marker="x")
        
    plt.show()
```

```python

#training loop
n_iter = 5
n_clusters=2


X_,Y_ = np.mgrid[-10:25:0.5 , -5:25:0.5 ]
print(X_)
pos = np.array([X_.flatten(),Y_.flatten()]).T
print(pos)


phi,mu,cov = init_params(X_train)

for i in range(n_iter):
    print(i)
    
    #plot current decision boundary
    plot_boundary(mu,cov,pos,X_train)
    
    
    #e step
    w = E_step(phi,mu,cov,n_clusters,X_train)
    
    #M step
    
    phi,mu,cov = M_step(phi,mu,cov,n_clusters,X_train,w)
    
    
    
```
