## Train Test Split

Because the data is ordered, we need to randomly sample points in the data in order to end up with a balanced test and train dataset. to do this we will sample from a random uniform law between 0 and 1 and filter those that come above .80 and store their indexes, this will serve as a sort of random sampling function

```python
N = X.shape[0]
I = X.shape[1]

split_ratio = 0.8
train_indexes = np.random.uniform(size=(I))  < split_ratio #returns boolean array with wether or not to sample this point

X_train = X[:,train_indexes]
y_train = y[train_indexes]

print(X_train.shape)

test_indexes = np.invert(train_indexes) #inverts boolean array, equivalent to NOT() operator

X_test = X[:,test_indexes]
y_test = y[test_indexes]

print(X_test.shape)
```

# Discriminative Model: Logistic Regression

Discriminative models which are also reffered as __conditional models__ model the dependence of unobserved (target) variables $y$ on observed (features) variables $X$. When take within a probabilistic framewrok this is effectively like modeling the conditional propability $ P(y|X) $ or the probability that a certain sample belongs to the target class y conditioned on the presence of the features $X$

One disadvantage with these is that it is not possible to sample data from its original distribution or in other words __generated new data points__. But, for tasks like classification and regression this fact is not important.


## Logistic regression

for logistic regression we take the case of binary classification where $y \in \{0,1\} $

An approach to this problem would be to find a function __$h_{\theta}(x) \in [0,1] $__ that essentially gives the propability of a given point __$X$__ belonging to one class versus another.

Linear regression is not exactly suited for classification, to counter this weakness we still build a linear model but use the logistic function (sigmoid function) where:

$$ h_{\theta}(X) = g(\theta^TX) = \frac{1}{1+ \exp{-\theta^TX}} $$

Making the assumption that this hypothesis function estimates the probability of a point belonging to a class, then we can write

$$ P(y=1|X;\theta) = h_{\theta}(X)$$
And
$$ P(y=0|X;\theta) = 1 - h_{\theta}(X)$$

Compacting the two expressions above we end up with

$$ P(y|X;\theta) = (h_{\theta}(X)^y)(1-h_{\theta}(X))^{(1-y)} $$

In order to optimize our algorithm, we will need to __maximize the  conditional likelyhood__. The likelyhood is defined as :
$$ L(\theta) = P(y|x;\theta) = \prod_{i=1}^{I} p(y^{(i)}|x^{(i)};\theta) = \prod_{i=1}^{I} (h_{\theta}(X^{(i)}))^y(1-h_{\theta}(X^{(i)}))^{(1-y)}   $$

The problem with computing the likelyhood, especially where out dataset becomes large is that this value will tend to go towards 0 and cause problems in computers like __underflow__. To counter this issue we will prefer to take the logarithm of the likelyhood and instead maximize the conditional log-lieklyhood. this gives us the following final expression:

$$ l(\theta) = log(L(\theta)) =  \sum_{i=1}^{I} y^{(i)}log((h_{\theta}(X^{(i)}))) + (1-y^{(i)}) log((1-h_{\theta}(X^{(i)}))) $$


In order to maximize $l(\theta)$ we can use the __gradient ascent algorithm__ which is extremely similar to gradient descent.

$$ \theta \leftarrow \theta + \alpha \nabla_{\theta} l(\theta) $$

here we consider $\theta$ as a column vector but we can write the expression for a single partial derivative as:

$$ \frac{\partial l(\theta)}{\partial \theta_n} = \sum_{i=1}^{I} (y^{(i)} - h_{\theta}(X^{(i)}) x_n^{(i)} $$

```python
def sigmoid(X,params):
    """ sigmoid function that will be the principal function of our model
        takes in parameters that we will multiply by the features
        returns: scalar corresponding to the predictions of our model between 0 and 1
    """
    return (1.0/(1.0 + np.exp(-1*np.dot(np.transpose(params),X))  ))
```

```python
def forward_pass(X,params):
    """ performs whole forward pass through our model
        returns: vector of predictions
    """
    return np.array([sigmoid(x_sample,params) for x_sample in np.transpose(X)])
```

```python
def gradient_ascent(params,X,y,yhat):
    """ Calculates one loop of the gradient ascent algorithm
        maximizes log-likelyhood for our logistic regression model
        returns: vector of updated parameters
    """
    alpha = 0.01
    
    grads = (np.sum( (y-yhat)*X ,axis=1))
    
    params += alpha * grads
    return params,grads
```

```python
def sse(y,yhat):
    return np.sum(np.power((yhat-y),2))/2.0
```

```python
def predict_log_reg(X,params):
    
    preds = forward_pass(X,params)
    
    return np.array([0.0 if pred < 0.5 else 1.0 for pred in preds ])
```

```python
params = np.random.normal(0,1,N)

n_epochs = 100

err = np.zeros((n_epochs,1))
param_tab = np.zeros((n_epochs,N))
grads_tab = np.zeros((n_epochs,N))

for epoch in range(n_epochs):
    yhat = forward_pass(X_train,params)
    
    err[epoch] = sse(y_train,yhat)
    
    params,grads = gradient_ascent(params,X_train,y_train,yhat)
    
    param_tab[epoch,:] = params
    grads_tab[epoch,:] = grads

    
%matplotlib inline

plt.figure()
plt.title("SSE")
plt.plot(err)

plt.show()

plt.figure()
plt.title("Parameters")
plt.plot(param_tab[:,0])
plt.plot(param_tab[:,1])

plt.show()


plt.figure()
plt.title("Gradients")
plt.plot(grads_tab[:,0])
plt.plot(grads_tab[:,1])
    
```

# Generative Model: Gaussian Discriminant Analysis (GDA)


Sources:
- http://web4.cs.ucl.ac.uk/staff/D.Barber/textbook/090113.pdf#page=425 : bayesian reasoning and machine learning book


Opposite to discriminative models we have __generative models__, these models are generative as they model the joint probability distribution $ P(X,y)$. From this distribution, we can both generate new samples or classify a given sample. This sort of model is called a  <a href="https://www.wikiwand.com/en/Mixture_model"> __mixture model__ </a>, because we use a combination of components to predict our total distribution.

In order to perform classification with the joint propability we need to make use of Bayes Rule : https://www.wikiwand.com/en/Bayes%27_theorem

with this rule we can go from a joint probability form to a conditional probability form.

### Bayes Rule:
 $$ P(X,y) = P(X|y)P(y) = P(y|X)P(X) $$
 
 
 
 $$ P(y|x)= \frac{P(X|y)P(y) }{P(X)} $$

Because we are doing binary classification meaning that we have two classes. We will be considering both $y = \{0,1\} $, because the $P(x)$ term will be the same for both of these, we can remove it from our calculation and simplify the expression to:

$$ P(y|x)= P(X|y)P(y)  $$


### Multivariate Normal Distribution

The multivariate normal distribution is the generalization of the normal distribution to N dimensions in space. it is parametrized by a __mean vector__ $\mu \in {\Bbb R}^N $ and a __covariance matrix__ $\Sigma \in{\Bbb R}^{N\times N} $ where $\Sigma \ge 0 $. The density is given by the expression:

$$ P(x;\mu,\Sigma) = \frac{1}{(\pi)^{\frac{n}{2}} |\Sigma|^{\frac{1}{2}} } \exp(-\frac{1}{2} (x-\mu)^T \Sigma^{-1} (x-\mu))   $$

### Maximum Likelyhood Estimation (MLE)
We optimize our model not through an iterative algorithm like gradient descent but through another method commonly used in statistics and known as __Maximum Likelyhood Explanation__. This way we can directly maximize the likelyhood using only one iteration of code instead of progressively tuning our parameters, 


```python
class GDA():
    
    def __init__(self,phi,mu0,mu1,cov):
        self.phi = phi
        self.mu0 = mu0
        self.mu1 = mu1
        self.cov = cov
        
        self.N = self.cov.shape[0]
    
    
    def get_densities(self,X):
        """ calculates the probability density for a given point for each class in our model
            returns vector of probabilites with index corresponding to class number
        """
        
        proba_0 = (1/(np.power(np.pi,self.N) * np.power(np.linalg.det(self.cov),1/2) )) \
                    *  np.exp((-1/2)* np.dot(np.dot(np.transpose(X-self.mu0), np.linalg.inv(self.cov) ), (X-self.mu0) ) )
        
        proba_1 = (1/(np.power(np.pi,self.N) * np.power(np.linalg.det(self.cov),1/2) )) \
                    *  np.exp((-1/2)* np.dot(np.dot(np.transpose(X-self.mu1), np.linalg.inv(self.cov) ), (X-self.mu1) ) )
        
        
        return np.array([proba_0,proba_1])
    
    def predict_probas(self,X):
        """ applies the multivariate normal density function 
            to the given features along with the trained parameters 
            in order to get the probability of a given point belonging to a certain class
            
            returns: matrix of samples and probas, need to take argmax later
        """
        
        return np.argmax(self.get_densities(X) * np.array([ (1-self.phi),self.phi]) )
        
        

    
```

```python
def get_covariance(X,y,mu0,mu1,I):
    """ calculates the covariance matrix of our data distribution. 
        here, we make the asumption that the covariance is the same
        for both distributions
    """
    #n,n matrix
    cov = np.zeros((X.shape[0],X.shape[0]))
    
    mus = [mu0.reshape((2,1)),mu1.reshape((2,1))]
    
    for x,label in zip(X.T,y):
        
        x = x.reshape((2,1))
        cov += np.dot((x-mus[label]),np.transpose(x-mus[label]))
        
    return (cov/I)
```

```python
def MLE(X,y):
    """ Calculates the parameters of our model from the data that we give it 
        returns the various parameters of our model like phi, mu and Sigma
    """
    I = X.shape[1]
    N = X.shape[0]
    
    #corresponds to likelyhood of belonging to a certain class
    phi = np.sum(y)/I
    
    mu0 = np.sum(X[:,y==0],axis=1)/(np.sum(abs(1-y) )) 
    
    mu1 =  np.sum(X[:,y==1],axis=1)/(np.sum(y))
    
    cov = get_covariance(X,y,mu0,mu1,I)
    
    print(mu0,mu1,cov)
    
    %matplotlib inline
    plt.scatter(X[0,:],X[1,:],c=y)
    plt.scatter(mu0[0],mu0[1],c="red",marker="x",s=75)
    plt.scatter(mu1[0],mu1[1],c="green", marker ="x",s=75)
    
    model = GDA(phi,mu0,mu1,cov)
    
    return model
    
gda_model = MLE(X_train,y_train)
```

## Predicting points

Now that we have trained our mixture model through MLE we can now both __generate and classify__ points. We do this by looking at the predictions made from each of our sub-models and take the prediction from which the classification is the highest.

This is the same as taking $$ argmax_y P(Y|X) =  argmax_y P(X|Y) P(Y)  $$ 

```python
def confusion_matrix(y,y_hat):
    
    mat = np.zeros((2,2))
    
    for true , pred in zip(y,y_hat):
        
        #true positive
        if pred == 1 and pred == true:
            mat[0,0] += 1
        #true negative
        if pred == 0 and pred == true:
            mat[1,1] += 1
        
        #false positive
        if pred == 1 and pred != true:
            mat[1,0] += 1
        #false negative
        if pred == 0 and pred != true:
            mat[0,1] += 1
    
    return mat
```
