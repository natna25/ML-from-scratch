## The problem

The goal of this lab in the first part is to build an adding machine for a sequence of 8 bits using a vanilla Elman RNN. With a sequence like this:
$$ 1011 0101 \rightarrow 5 <=> \sum seq$$

In this sort of problem, our input is a sequence of bits X and the output is only measured after the __entire__ sequence has been passed through the network

## Generating the Data

```python
def generate_sequences(n = 50 ):
    """ generates sequences of 8 bits with random 0 and ones and 
        Returns both the features (sequence) and labels (sum)
    """
    X = np.random.randint(0,2,size=(n,8)) #features
    y = np.sum(X,axis=1) # labels, outputs
    
    return X,y

X,y = generate_sequences(50)
print(X[:5],y[:5])
```

### Problem simplifications 
Due to the simplicity of the problem at hand, there are a few simplifications that we can make to the standard RNN architecture that will make our problem significantly easier

1) No bias, Because we are adding numbers together and the previous one can be zero, we do not need to have a bias in our model. Take the case were the sequence is all zeros, in this case we need our model to pass by the origin.

2) The desired output is linear, we do not need to apply a non-linear activation function in between neurons 

## making our model
 Because of the simplifications that we noted earlier, we will have a simple model that will only have two parameters

## Forward propagation

The forward propagation is the first phase that we need to do before backpropagation, we will pass in our sequence of numbers and return the final result. From this forward pass, we can then calculate our error  and proceed to the backpropagation phase.

The formula for this is the following:

## $$ \{ 
    \begin{array}{ll}
        f_t = V_f f_{t-1} + V_x x_t \\
        \hat{y} = f_T
    \end{array}  $$
    
We can notice that there are no biases here due to the simplifications that we could apply to our problem, there is also no activation function

```python
def forward_prop(X,Vf,Vx):
    """ function for the forward propagation phase of our RNN
        returns the final results after the whole sequence has been treated
    """
    #ft_1 = np.zeros(X.shape[0]) # ft-1, previous output
    ft = np.zeros(X.shape[0]) #current output we c
    
    fi = np.zeros(X.shape) #to store intermediate values of Ft
    
    for t in range(X.shape[1]):
        
        ft = Vf * ft + Vx * X[:,t]
        
        fi[:,t] = ft
        
    
    
    
    return ft,fi
        
```

## Back Propagation

Backprop is the second phase of the training process of a machine learning algorithm. After performing the forward pass and calculating the error. this backpropagation phase is achieved using the gradient descent algorithm whose expression in our case is the following.

## $$ \{ 
    \begin{array}{ll}
        V_x^{itera} = V_x^{itera-1} - \alpha \frac{\partial E}{\partial V_x} \\
        V_f^{itera} = V_f^{itera-1} - \alpha \frac{\partial E}{\partial V_f}
    \end{array}  $$
    
The values that we do not know are the partial derivatives of the error with respects to the parameters $V_f$ and $V_x$ . These gradients tell us __how to modify these parameters in order to reduce the error__

Here are the values of the partial derivatives that we will implement

## $$ \frac{\partial E}{\partial V_x} = \sum_{t=1}^T [ \sum_{i=1}^I (\hat{y}^{(i)} - y^{(i)} ) x_t^{(i)} ] V_f^{T-t} $$

## $$ \frac{\partial E}{\partial V_f} = \sum_{t=1}^T [ \sum_{i=1}^I (\hat{y}^{(i)} - y^{(i)} ) f_{t-1}^{(i)} ] V_f^{T-t}  $$




```python
def backprop(X, y,yhat,Vf,Vx,fi):
    """ vanilla backpropagation function
        this is called at each batch or foward pass
        updates the values of the parameters and returns Vf and Vx
    """
    #constants
    alpha = 0.0001
    
    #calculate the error
    error = yhat - y
    
    dVf = 0
    dVx = 0
    
    T = X.shape[1]
    
    for t in range(T):
        if t ==0:
            dVf += (np.sum(error * 0,axis=0) * (Vf**(T-t) ) ) 
            dVx += (np.sum(error * X[:,t]) * (Vf**(T-t) ) )
        else:
            dVf += (np.sum(error * fi[:,t-1],axis=0) * (Vf**(T-t) ) )
            dVx += (np.sum(error * X[:,t]) * (Vf**(T-t) ) )
        
    Vf -= (alpha * dVf)
    Vx -= (alpha * dVx)
    
    
    return Vf,Vx,dVf,dVx
    
```

```python
def sse(y,yhat):
    return np.sum(np.power((yhat-y),2))/2.0
```

```python
#training loop

n_epochs = 500

#initialize the parameters randomly with a normal distribution
Vf = np.random.normal(1,1) #normal with mean 1 and std 1
Vx = np.random.normal(1,1)

params = np.zeros((n_epochs,2))

X,y = generate_sequences(50)

#arrays to store error and gradients over time
err = []
grads = np.zeros((n_epochs,2))

# training loop
for epoch in range(n_epochs):

    #forward prop phase gets us predictions and we store intermediate values
    yhat,fi = forward_prop(X,Vf,Vx)
    
    err.append(sse(y,yhat))
    
    #backprop, we update the parameters and look at the gradients 
    Vf,Vx,dVf,dVx = backprop(X,y,yhat,Vf,Vx,fi)
    
    params[epoch,0] = Vf
    params[epoch,1] = Vx
    
    #print(Vf,Vx)
    grads[epoch,0] = dVf
    grads[epoch,1] = dVx

    

plt.figure()    
plt.plot(err)
plt.title("Error")
plt.show()

plt.figure()
plt.plot(grads[:,0])
plt.plot(grads[:,1])
plt.title("Gradients")
plt.show()

plt.figure()
plt.plot(params[:,0],label="Vf")
plt.plot(params[:,1],label="Vx")
plt.legend()
plt.title("model parameters")
plt.show()
```

## Vanishing and exploding gradient problem

RNNs have a problem dealing with long term dependencies, this is due to the way that the gradient is propagated throught the layers (steps in the sequence). The longer the sequence is, the more matrix multiplications we apply. If the parameters are less than than 1 then the gradient will go towards 0 and undeflow because it gets so small that it cannot be represented as a float anymore. Inversly, if the parameter is greater than 1 then this will tend towards infinity. In the first case we say that the gradients __vanish__, in the second case we say that they __explode__.

To deal with this, we need to tweak our gradient descent algorithm to be more robust to long sequences and a possible high gradient. We will explore two different methods of doing this, __resilient propagation__ and __gradient clipping__

## Resilient Propagation

This algorithm proposed by two researchers (Martin Riedmiller and Heinrich Braun) in 1992 aims at solving the gradient vanishing and explosion problem. Instead of using the value of the gradient to guide our steps we look at the sign of these gradients and how they may change over time. Then, we take only a small step of a predefined (controlled) size in a direction. There is also a functionality analogous to a dynamic learning rate in order to converge faster.



## $$ \{ 
    \begin{array}{ll}
        V_x^{itera} = V_x^{itera-1} - Sign(\frac{\partial E}{\partial V_x}) \Delta_x \\
        V_f^{itera} = V_f^{itera-1} - Sign(\frac{\partial E}{\partial V_f}) \Delta_f
    \end{array}  $$
    
    
Where

## $$ \eta_p = 1.2 ;  
   \eta_n = 0.5 ;
   \Delta_x^{init}=0.001$$

## $$ 
    \Delta_x = 
    \{ 
    \begin{array}{ll}
        \Delta_{x}^{(itera-1)} \eta_p \; if \; Sign(\frac{\partial E}{\partial V_x})_{(itera-1)} == Sign(\frac{\partial E}{\partial V_x})_{(itera)}  \\
        \Delta_{x}^{(itera-1)} \eta_n \; otherwise
    \end{array}  $$



```python
def resilient_prop(X, y,yhat,Vf,Vx,fi,prev_dVf,prev_dVx,delta_f,delta_x):
    """ implementation of the resilient propagation algorithm 
        as explained above
    """
    #constants
    eta_p = 1.2
    eta_n = 0.5
    
    
    #calculate the error
    error = yhat - y
    
    #gradients
    dVf = 0
    dVx = 0
    
    min_delta = 1e-6
    max_delta = 20
    
    T = X.shape[1]
    
    for t in range(T):
        if t ==0:
            dVf += np.sum(error * 0,axis=0) * (Vf**(T-t) ) 
            dVx += np.sum(error * X[:,t]) * (Vf**(T-t) ) 
        else:
            dVf += np.sum(error * fi[:,t-1],axis=0) * (Vf**(T-t) ) 
            dVx += np.sum(error * X[:,t]) * (Vf**(T-t) ) 
        
        
    if np.sign(prev_dVf) == np.sign(dVf):
        delta_f *= eta_p
    else:
        delta_f *= eta_n
        
    if np.sign(prev_dVx) == np.sign(dVx):
        delta_x *= eta_p
    else:
        delta_x *= eta_n
        
    Vf -= np.sign(dVf) * delta_f
    Vx -= np.sign(dVx) * delta_x
        
    
    
    return Vf,Vx,dVf,dVx,delta_f,delta_x
    
    
```

```python
n_epochs = 500
Vf = np.random.normal(1,1)
Vx = np.random.normal(1,1)

params = np.zeros((n_epochs,2))



err = []
grads = np.zeros((n_epochs,2))

dVf = 0
dVx = 0

delta_f = 0.001
delta_x = 0.001

for epoch in range(n_epochs):

    #forward prop phase gets us predictions and we store intermediate values
    yhat,fi = forward_prop(X,Vf,Vx)
    
    err.append(sse(y,yhat))
    
    #backprop, we update the parameters and look at the gradients 
    Vf,Vx,dVf,dVx,delta_f,delta_x = resilient_prop(X,y,yhat,Vf,Vx,fi,dVf,dVx,delta_f,delta_x)
    
    params[epoch,0] = Vf
    params[epoch,1] = Vx
    
    #if epoch % 5 == 0:
        #print(delta_f,delta_x)
        #print("grad:",dVf,dVx)
        #print("params:",dVf,Vx)
    
    grads[epoch,0] = dVf
    grads[epoch,1] = dVx
  
    
plt.figure()    
plt.plot(err)
plt.title("Error")
plt.show()

plt.figure()
plt.plot(grads[:,0],label="dVf")
plt.plot(grads[:,1],label="dVx")
plt.legend()
plt.title("Gradients")
plt.show()

plt.figure()
plt.plot(params[:,0],label="Vf")
plt.plot(params[:,1],label="Vx")
plt.legend()
plt.title("model parameters")
plt.show()
```

## Gradient Clipping

Another method that one can do is called gradient clipping in order to prevent gradient vanishing and explosion. The idea here is to monitor the values of the gradients and to 'clip' them if they get above a certain value. This clipping can prevent gradient explosion but is not imune to gradient vanishing. 
Here is a visualization on how gradient clipping helps limit the step size when the gradient spikes up
<img src="gradient_clipping.png">

source: <a href="http://www.deeplearningbook.org/contents/rnn.html"> Deep Learning book </a>, Ian Goodfellow

```python
def gradient_clipping(X,y,yhat,Vf,Vx,fi):
    
    #constants
    alpha = 0.001
    eta = 1.0
    
    #calculate the error
    error = yhat - y
    
    dVf = 0
    dVx = 0
    
    T = X.shape[1]
    
    for t in range(T):
        
        if t ==0:
            dVf += (np.sum(error * 0,axis=0) * (Vf**(T-t) ) ) 
            dVx += (np.sum(error * X[:,t]) * (Vf**(T-t) ) )
        else:
            dVf += (np.sum(error * fi[:,t-1],axis=0) * (Vf**(T-t) ) )
            dVx += (np.sum(error * X[:,t]) * (Vf**(T-t) ) )
            
    
    
    if np.linalg.norm(dVf) > eta:
        Vf -= (alpha * (eta*(dVf/np.linalg.norm(dVf))))
    else:
        Vf -= (alpha * dVf)
        
    if np.linalg.norm(dVx) > eta:
        Vx -= (alpha * (eta*(dVx/np.linalg.norm(dVx))))
    else:
        Vx -= (alpha * dVx)
    
    
  
    
    return Vf,Vx,dVf,dVx
```

```python
n_epochs = 1500
Vf = np.random.normal(1,1)
Vx = np.random.normal(1,1)

params = np.zeros((n_epochs,2))

err = []
grads = np.zeros((n_epochs,2))

dVf = 0
dVx = 0

for epoch in range(n_epochs):

    #forward prop phase gets us predictions and we store intermediate values
    yhat,fi = forward_prop(X,Vf,Vx)
    
    err.append(sse(y,yhat))
    
    #backprop, we update the parameters and look at the gradients 
    Vf,Vx,dVf,dVx = gradient_clipping(X,y,yhat,Vf,Vx,fi)
    
    params[epoch,0] = Vf
    params[epoch,1] = Vx
    
    grads[epoch,0] = dVf
    grads[epoch,1] = dVx
    
    
plt.figure()    
plt.plot(err)
plt.title("Error")
plt.show()

plt.figure()
plt.plot(grads[:,0],label="dVf")
plt.plot(grads[:,1],label="dVx")
plt.legend()
plt.title("Gradients")
plt.show()

plt.figure()
plt.plot(params[:,0],label="Vf")
plt.plot(params[:,1],label="Vx")
plt.legend()
plt.title("model parameters")
plt.show()
```
